```python
#f(x) = x^2+exp(−1/(100(x−1))^2) -1
#f'(x) = e^(-1/(10000 (x-1)^2))/(5000 ( x-1)^3) + 2 x 
def derivative_of_fun(x):
    derv = 2*x + (math.exp(-1/(10000*(x-1)**2)))/(5000*(x-1)**3)
    #derv = 5*x**4
    return derv


def update_xvalue(x, lr):
    '''
    Vectorized Gradient Descent

    Features:(200, 3)
    Labels: (200, 1)
    Weights:(3, 1)
    '''
    #new_x = x
    
    #gradient = first derivative of f(x) = x**2
    d = derivative_of_fun(x)

    #4 - Multiply the gradient by our learning rate
    gradient = lr*d

    #5 - Subtract from currrent x value
    new_x = x - gradient

    return new_x
```

```python
def execute_gradient(x, lr):
    itr = 0
    x_updated = x
    while(True):
        itr = itr + 1
        x_current = x_updated
        print(x_current)
        x_updated = update_xvalue(x_current ,lr)
        
        if(np.allclose(x_current,x_updated)): #Until its previous version and current version are not close, we iterate
            break
            
    return x_updated,itr
    
```

```python
x_in = float(2.0)
out_x , n_iter = execute_gradient(x_in,0.0001)
print(out_x ,',', n_iter)
```
