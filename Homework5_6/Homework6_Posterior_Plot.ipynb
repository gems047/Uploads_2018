```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
```

```python



mu1 = 1
var1 = 2
sigma1 = math.sqrt(var1)

mu2 = 3
var2 = 1
sigma2 = math.sqrt(var2)

x1 = np.linspace(mu1 - 3*sigma1, mu1 + 3*sigma1, 100)
plt.plot(x1,mlab.normpdf(x1, mu1, sigma1))


x2 = np.linspace(mu2 - 3*sigma2, mu2 + 3*sigma2, 100)
plt.plot(x2,mlab.normpdf(x2, mu2, sigma2))
print(x2)

plt.xlabel('x')
plt.ylabel('Probability density - f(x|mu,sigma^2)')
plt.show()
```

```python
def g1_posterior(x):
    g1 = -1.777-np.power((x-1),2)/4
    #g1 = -1.777-(x-1)**2/4
    return g1

def g2_posterior(x):
    g2 = -1.84-(x-3)**2/2
    return g2

#Posterior probabilities P(w1/x) as ln[P(w1/x)]
g1_out = g1_posterior(x1)
g2_out = g2_posterior(x2)
#print(g1_out)


# c = 2 + np.array([1, 2, 3])
# print(c)

#Plot posterior prob against x values

plt.plot(x1,g1_out)
plt.plot(x2,g2_out)
plt.xlabel('x')
plt.ylabel('posterior-ln[P(w/x)]')

plt.show()

```
